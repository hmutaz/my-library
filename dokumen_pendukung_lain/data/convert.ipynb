```python
import pandas as pd
from rdflib import Graph, Literal, URIRef, Namespace
from rdflib.namespace import XSD
from rdflib.extras.describer import Describer
```

```python
df = pd.read_csv('seattle_inventory.csv')
```

```python
len(df)
```

```python
df = df.drop_duplicates(subset=["Title", "ItemType", "ItemCollection", "ItemLocation"])
```

```python
df.loc[df['Title'] == 'The 5 essentials : using your inborn resources to create a fulfilling life / Bob Deutsch, PhD ; with Lou Aronica.']
```

```python
df = df.drop(columns=["ItemType", "FloatingItem"])
```

```python
df.head(5)
```

```python
g = Graph()
lib = Namespace('http://www.semanticweb.org/m13/ontologies/2024/4/library/')
g.bind('lib', lib)
d = Describer(Graph)
d.about(lib+"123")
```

```python
for index, row in df.iterrows():
    g.add((URIRef(lib+str(row['BibNum'])), URIRef(lib+'title'), Literal(row['Title'], datatype=XSD.string)))
    g.add((URIRef(lib+str(row['BibNum'])), URIRef(lib+'author'), Literal(row['Author'], datatype=XSD.string)))
    g.add((URIRef(lib+str(row['BibNum'])), URIRef(lib+'isbn'), Literal(row['ISBN'], datatype=XSD.string)))
    g.add((URIRef(lib+str(row['BibNum'])), URIRef(lib+'year'), Literal(row['PublicationYear'], datatype=XSD.string)))
    g.add((URIRef(lib+str(row['BibNum'])), URIRef(lib+'publisher'), Literal(row['Publisher'], datatype=XSD.string)))
    g.add((URIRef(lib+str(row['BibNum'])), URIRef(lib+'subjects'), Literal(row['Subjects'], datatype=XSD.string)))
    g.add((URIRef(lib+str(row['BibNum'])), URIRef(lib+'collection'), Literal(row['ItemCollection'], datatype=XSD.string)))
    g.add((URIRef(lib+str(row['BibNum'])), URIRef(lib+'location'), Literal(row['ItemLocation'], datatype=XSD.string)))
    g.add((URIRef(lib+str(row['BibNum'])), URIRef(lib+'reportdate'), Literal(row['ReportDate'], datatype=XSD.string)))
    g.add((URIRef(lib+str(row['BibNum'])), URIRef(lib+'count'), Literal(row['ItemCount'], datatype=XSD.string)))
```

```python
print(g.serialize(format='xml'))
```
